# LOAD DATA (macmorpho-v3)

```python
def load_data_to_list(path):
  X = []
  Y = []

  with open(path, 'r') as f:
    lines = f.readlines()

  for line in lines:
    result = line.split(' ')
    X.append([l.split('_')[0].lower() for l in result])
    Y.append([l.split('_')[1].lower().replace('\n','').replace('+','').replace('-','') for l in result])

  return X, Y
```

```python
def get_all_tags(vec):
  dict_tags = {}
  
  for tags in vec:
    for tag in tags:
      if tag not in dict_tags:
        dict_tags[tag] = 1
      else:
        dict_tags[tag] += 1
    
  return dict_tags
```

```python
TRAIN_SENTENCES, TRAIN_TAGS = load_data_to_list(FILE_DATA_TRAIN)
TEST_SENTENCES, TEST_TAGS = load_data_to_list(FILE_DATA_TEST)

#JOIN
ALL_SENTENCES = []
ALL_SENTENCES.extend(TRAIN_SENTENCES)
ALL_SENTENCES.extend(TEST_SENTENCES)

print(TRAIN_SENTENCES[0])
print(TEST_SENTENCES[0])
print(TRAIN_TAGS[0])
```

```python
TRAIN_TAGS_COUNT = get_all_tags(TRAIN_TAGS)
TEST_TAGS_COUNT = get_all_tags(TEST_TAGS)
ALL_TAGS_COUNT = {k: v + TEST_TAGS_COUNT[k] for k,v in TRAIN_TAGS_COUNT.items()}

NUMBER_TAGS = len(ALL_TAGS_COUNT)
print('ALL_TAGS: ', NUMBER_TAGS)
print(TEST_TAGS_COUNT)
```

# Load Embedding

```python
def load_word_embedding(word_index, dimension_embedding):
    dic_embedding = {}
    
    file_embedding = open(os.path.join('/home/daniel/Downloads', "skip_s{}.txt".format(dimension_embedding)), encoding='utf-8')
    
    #Carregando os WordEmbeddings para um dicionario
    for line in tqdm(file_embedding):
        values = line.rstrip().rsplit(' ')
        word = values[0]
        coefficient = np.asarray(values[1:], dtype='float32')
        dic_embedding[word] = coefficient
            
    file_embedding.close()
    
    #Cria a matrix com em que serão carregados os embeddings
    matrix_embedding = np.zeros((VOCABULARY_SIZE, dimension_embedding))
    
    for word, i in word_index.items():
        if i >= VOCABULARY_SIZE:
            continue
        
        vector_embedding = dic_embedding.get(word)
        
        if vector_embedding is not None:
            matrix_embedding[i] = vector_embedding

    return matrix_embedding
```

```python
MATRIX_EMBEDDING = load_word_embedding(word_tokenizer.word_index, EMBEDDING_SIZE)
```

#LSTM - Model

```python
def ignore_class_accuracy(to_ignore=0):
    def ignore_accuracy(y_true, y_pred):
        y_true_class = K.argmax(y_true, axis=-1)
        y_pred_class = K.argmax(y_pred, axis=-1)
 
        ignore_mask = K.cast(K.not_equal(y_pred_class, to_ignore), 'int32')
        matches = K.cast(K.equal(y_true_class, y_pred_class), 'int32') * ignore_mask
        accuracy = K.sum(matches) / K.maximum(K.sum(ignore_mask), 1)
        return accuracy
    return ignore_accuracy
```

```python
# create architecture

lstm_model = Sequential()
lstm_model.add(Embedding(input_dim     = VOCABULARY_SIZE,         # vocabulary size - number of unique words in data
                         output_dim    = EMBEDDING_SIZE,          # length of vector with which each word is represented
                         input_length  = MAX_WORDS_SENTENCES,          # length of input sequence
                         weights       = [MATRIX_EMBEDDING],     # word embedding matrix
                         trainable     = True,                     # True - update embeddings_weight matrix
                         mask_zero     = True                     #https://www.tensorflow.org/guide/keras/masking_and_padding || https://data-flair.training/blogs/masking-and-padding-in-keras/
))

#lstm_model.add(Bidirectional(LSTM(64, return_sequences=True, recurrent_dropout=0.1)))
lstm_model.add(LSTM(32, return_sequences=True))

#lstm_model.add(Dense(500, activation='sigmoid'))
#lstm_model.add(Dropout(0.50))

#lstm_model.add(Dense(500, activation='ReLU'))
#lstm_model.add(Dropout(0.50))

lstm_model.add(TimeDistributed(Dense(NUM_CLASSES, activation='softmax')))
```

```python
lstm_model.summary()
```

```python
lstm_model.compile(loss      =  'categorical_crossentropy',
                   optimizer =  'adam',
                   metrics   =  ['acc'])
                   #sample_weight_mode="temporal")
```

```python
cc = {i: j for i,j in class_weights_values.items() if i < 2}
```

#RNN MODEL

```python
# create architecture

rnn_model = Sequential()

# create embedding layer - usually the first layer in text problems
rnn_model.add(Embedding(input_dim     =  VOCABULARY_SIZE,         # vocabulary size - number of unique words in data
                        output_dim    =  EMBEDDING_SIZE,          # length of vector with which each word is represented
                        input_length  =  MAX_WORDS_SENTENCES,          # length of input sequence
                        weights       =  [MATRIX_EMBEDDING],
                        trainable     =  False,                    # False - don't update the embeddings
                        mask_zero     =  False
))

# add an RNN layer which contains 64 RNN cells
rnn_model.add(SimpleRNN(64, #64 
              return_sequences=True  # True - return whole sequence; False - return single output of the end of the sequence
))

# add time distributed (output at each sequence) layer
rnn_model.add(TimeDistributed(Dense(NUM_CLASSES, activation='softmax')))
```

```python
rnn_model.compile(loss      =  'categorical_crossentropy',
                  optimizer =  'adam',
                  metrics   =  ['accuracy', ignore_class_accuracy(0)])
```

```python
rnn_model.summary()
```

```python
rnn_training = rnn_model.fit(X_TRAIN, Y_TRAIN, batch_size=512, epochs=5, validation_data=(X_TEST, Y_TEST))
```

```python
plt.plot(rnn_training.history['accuracy'])
plt.plot(rnn_training.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc="lower right")
plt.show()
```

```python
def logits_to_tokens(sequences, index):
    token_sequences = []
    for categorical_sequence in sequences:
        token_sequence = []
        
        for categorical in categorical_sequence:          
            if np.argmax(categorical) != 0:
                token_sequence.append(index[np.argmax(categorical)])
                
        token_sequences.append(token_sequence)
    return token_sequences

  #https://nlpforhackers.io/lstm-pos-tagger-keras/
```

```python
def remove_onehot(sequences):
    token_sequences = []
    for categorical_sequence in sequences:
        token_sequence = []
        
        for categorical in categorical_sequence:          
          if np.argmax(categorical) != 0:
            token_sequence.append(np.argmax(categorical))
        token_sequences.append(token_sequence)
 
    return token_sequences
```

```python
def to_categorical(sequences, categories):
    cat_sequences = []
    for s in sequences:
        cats = []
        for item in s:
            cats.append(np.zeros(categories))
            cats[-1][item] = 1.0
        cat_sequences.append(cats)
    return np.array(cat_sequences)
 
```
